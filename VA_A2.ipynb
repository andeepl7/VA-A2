# **Question 3**
Adding Continent column to the Dataframe

```python
continent_url='https://raw.githubusercontent.com/dbouquin/IS_608/master/NanosatDB_munging/Countries-Continents.csv'
continent_country = pd.read_csv(continent_url)
continent_country
```

```python
row1 = {"Continent": "Not_Available", "Country":"Not_Available"}
row2 = {"Continent": "Multi_nation", "Country": "multi_nation"}
row3 = {"Continent": "Asia", "Country": "Taiwan"}
row4 = {"Continent": "Asia", "Country": "Hong Kong"}
continent_country = continent_country.append(row1, ignore_index=True)
continent_country = continent_country.append(row2, ignore_index=True)
continent_country = continent_country.append(row3, ignore_index=True)
continent_country = continent_country.append(row4, ignore_index=True)
continent_country['Country'] = continent_country['Country'].replace(['Korea, South'], 'South Korea')
continent_country['Country'] = continent_country['Country'].replace(['Russian Federation'], 'Russia')
continent_country
```

```python
netflix_clone = netflix.copy()
```

```python
netflix_clone['country'] = netflix_clone['country'].replace(['United States'], 'US')
netflix_clone['country'] = netflix_clone['country'].replace([', South Korea'], 'South Korea')
netflix_clone['country'] = netflix_clone['country'].replace(['United Kingdom,'], 'United Kingdom')
netflix_clone['country'] = netflix_clone['country'].fillna("Not_Available")
```

```python
netflix_clone.loc[netflix_clone['country'].str.contains(','), 'country'] = 'multi_nation'
```

```python
list_values1 = netflix_clone.country.value_counts()
df1 = list_values1.to_frame()
```

```python
netflix_clone2 = pd.merge(netflix_clone, continent_country, how='inner', left_on="country", right_on = "Country")
netflix_clone2
```

```python
netflix_clone2['director'].str.strip()
netflix_clone2['title'].str.strip()
netflix_clone2['director']= netflix_clone2['director'].str.replace('\"','')
netflix_clone2['director']= netflix_clone2['director'].str.replace('-','')
netflix_clone2['director']= netflix_clone2['director'].str.replace('.','')
netflix_clone2['director']= netflix_clone2['director'].str.replace('!','')
netflix_clone2['director']= netflix_clone2['director'].str.replace(',','')
netflix_clone2['title']= netflix_clone2['title'].str.replace('-','')
netflix_clone2['title']= netflix_clone2['title'].str.replace('\"','')
netflix_clone2['title']= netflix_clone2['title'].str.replace('.','')
netflix_clone2['title']= netflix_clone2['title'].str.replace('!','')
netflix_clone2['title']= netflix_clone2['title'].str.replace(',','')
netflix_clone2
```

```python
list_value2 = netflix_clone2.country.value_counts()
df2 = list_value2.to_frame()
```

```python
pd.concat([df1,df2]).drop_duplicates(keep=False)
```

```python
continentplot = sns.catplot(data=netflix_clone2, x="Continent", kind="count", palette="rocket", edgecolor=".6", height=5, aspect = 2)
continentplot.set_xlabels('Continent', fontsize=15)
continentplot.set_ylabels('Count', fontsize=15)
```

# **Question 4**

```python
imdb_votes_url='https://github.com/andeepl7/VA-A2/blob/main/imdb_votes.xlsx?raw=true'
myfile = requests.get(imdb_votes_url)
dest = 'local-file.xlsx'
with open(dest, 'wb') as file:
    file.write(myfile.content)

imdb_votes = pd.read_excel(dest)
imdb_votes.head()
```

```python
imdb_movies_url='https://raw.githubusercontent.com/andeepl7/VA-A2/main/imdb_movies.csv'
imdb_movies = pd.read_csv(imdb_movies_url)
imdb_movies.head()
```

```python
imdb_database = pd.merge(imdb_movies, imdb_votes, how='inner', left_on="imdb_title_id", right_on = "imdb_title_id")
imdb_database
```

```python
imdb_database.title.duplicated().sum()
```

```python
imdb_database[imdb_database.title.duplicated() == True]
```

```python
imdb_database[imdb_database.duplicated(subset=['title','director'], keep=False)]
```

```python
df_imdb_clean = imdb_database.drop_duplicates(subset = ['title','director'],keep = 'last').reset_index(drop = True)

df_imdb_clean
```

# **Question 5**

```python
df_imdb_clean['director'].str.strip()
df_imdb_clean['title'].str.strip()
df_imdb_clean['director']= df_imdb_clean['director'].str.replace('\"','')
df_imdb_clean['director']= df_imdb_clean['director'].str.replace('!','')
df_imdb_clean['director']= df_imdb_clean['director'].str.replace('-','')
df_imdb_clean['director']= df_imdb_clean['director'].str.replace('.','')
df_imdb_clean['director']= df_imdb_clean['director'].str.replace(',','')
df_imdb_clean['title']= df_imdb_clean['title'].str.replace('\"','')
df_imdb_clean['title']= df_imdb_clean['title'].str.replace('!','')
df_imdb_clean['title']= df_imdb_clean['title'].str.replace('-','')
df_imdb_clean['title']= df_imdb_clean['title'].str.replace('.','')
df_imdb_clean['title']= df_imdb_clean['title'].str.replace(',','')
df_imdb_clean
```

```python
netflix_imdb = pd.merge(netflix_clone2, df_imdb_clean, how='inner', left_on=['title','director'], right_on = ['title','director'])
netflix_imdb
```
